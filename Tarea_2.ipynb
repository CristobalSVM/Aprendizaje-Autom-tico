# Tipo de variable

```python
convertir = {'id': int,
                'fecha': int,
                'sucursales': int,
                'cajeros': int,
                'tarjetas': int}  
  
df = df.astype(convertir)  
print(df.dtypes)
```

# Crear, modificar o eliminar columnas

```python
# modificar columnas fecha, activo, vista, plazo, prestamos interbancarios para tener unidades de millones de pesos
df['activo'] = df['activo']/1000000 
df['vista'] = df['vista']/1000000
df['plazo'] = df['plazo']/1000000
df['prestamos interbancarios'] = df['prestamos interbancarios']/1000000
df
```

```python
# Crear columna captacion = vista + plazo + prestamos interbancarios para tener las proporciones de estas mismas
captacion = ['vista', 'plazo', 'prestamos interbancarios']

df_captacion = df[captacion]
df['captacion'] = df_captacion.sum(axis=1)
df['% vista'] = df_captacion['vista']/df_captacion.sum(axis=1)
df['% plazo'] = df_captacion['plazo']/df_captacion.sum(axis=1)
df['% prestamos interbancarios'] = df_captacion['prestamos interbancarios']/df_captacion.sum(axis=1)

df
```

```python
# Eliminar columna id
df = df.drop(['id'],axis=1)
```
